```python
df = df.drop_duplicates()
```

```python
#removing link & property column

df.drop(columns=['link', 'property_id'], inplace = True)
```

```python
df.rename(columns= {'rate' : 'price_per_sqft'}, inplace=True)
```

```python
df['society'].value_counts()
```

```python
df['society'].isnull().sum()
```

```python
import re

df['society'] = df['society'].apply(lambda name: re.sub(r'(\d+\.\d+?\s★)', '', str(name)).strip()).str.split('/').str.get(0).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society'] = df['society'].str.replace('nan', 'independent')
```

```python
df.head()
```

```python
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request'].shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):

  if type(x) == float:
    return x

  else:

    if x[1] == 'Lac':

      return round(float(x[0])/100,2)

    else:

      return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',', '').str.strip().astype(float)
```

```python
df.head()
```

```python
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df.shape
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df.shape
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).str.strip().astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).str.strip().astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.strip().str.replace('No', '0')
```
